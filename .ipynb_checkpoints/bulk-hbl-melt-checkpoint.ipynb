```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import seaborn as sns
```

```python
bulk_melt = pd.read_csv('vol_plu_all.csv')
```

```python
bulk_meltCM = bulk_melt[bulk_melt['TECTONIC SETTING'] == 'CONVERGENT MARGIN']
vol = bulk_meltCM[bulk_meltCM['ROCK TYPE'] == 'VOL']
plu = bulk_meltCM[bulk_meltCM['ROCK TYPE'] == 'PLU']
```

```python
# bulk-melt
bulk_elements = ['SIO2(WT%)', 'TIO2(WT%)', 'AL2O3(WT%)',
                 'FEOT(WT%)', 'MGO(WT%)', 'CAO(WT%)', 'K2O(WT%)']
melt_elements = ['mSiO21', 'mSiO22', 'mSiO23', 'mSiO24', 'mTiO21', 'mTiO22',
                 'mFeO1', 'mFeO2', 'mMgO', 'mCaO1', 'mCaO2', 'mK2O1', 'mK2O2', 'mAl2O3']

for b in bulk_elements:
    for m in melt_elements:
        if b.lower()[:-6] in m.lower():
            fig, axs = plt.subplots(ncols=2, nrows=1, figsize=(10, 4))

            sns.kdeplot(x=vol[b], y=vol[m], shade=True, color='#E60000',
                        #                        scatter_kws={'alpha':0.4, 's':4,}, fit_reg = False, x_jitter = 0.5, y_jitter = 0.2,
                        ax=axs[0])
            axs[0].scatter(vol[b], vol[m], color='gray', alpha=.3, s=1)
            axs[0].plot(np.arange(0, 100), np.arange(0, 100), c='gray')
            axs[0].set_title(r'$Volcanic$')
#             axs[0].set_xlabel('Bulk ' + b)
#             axs[0].set_ylabel('Melt ' + b)

            sns.kdeplot(x=plu[b], y=plu[m], shade=True, color='#003153',
                        #                          scatter_kws={'alpha':0.4, 's':4,}, fit_reg = False, x_jitter = 0.5, y_jitter = 0.2,
                        ax=axs[1])
            axs[1].scatter(plu[b], plu[m], color='gray', alpha=.3, s=1)
            axs[1].plot(np.arange(0, 100), np.arange(0, 100), c='gray')
            axs[1].set_title(r'$Plutonic$')
            axs[1].set_xlabel('Bulk ' + b)
            axs[1].set_ylabel('Melt ' + b)

            if b == 'SIO2(WT%)':
                axs[0].set_xlim(40, 80)
                axs[0].set_ylim(40, 80)
                axs[0].set_xlabel(r'$Bulk$ ' + r'$SiO_2$')
                axs[0].set_ylabel(r'$Melt$ ' + r'$SiO_2$')

                axs[1].set_xlim(40, 80)
                axs[1].set_ylim(40, 80)
                axs[1].set_xlabel(r'$Bulk$ ' + r'$SiO_2$')
                axs[1].set_ylabel(r'$Melt$ ' + r'$SiO_2$')

            elif b == 'TIO2(WT%)':
                axs[0].set_xlim(0, 2)
                axs[0].set_ylim(0, 2)
                axs[0].set_xlabel(r'$Bulk$ ' + r'$TiO_2$')
                axs[0].set_ylabel(r'$Melt$ ' + r'$TiO_2$')

                axs[1].set_xlim(0, 2)
                axs[1].set_ylim(0, 2)
                axs[1].set_xlabel(r'$Bulk$ ' + r'$TiO_2$')
                axs[1].set_ylabel(r'$Melt$ ' + r'$TiO_2$')

            elif b == 'AL2O3(WT%)':
                axs[0].set_xlim(10, 23)
                axs[0].set_ylim(10, 22)
                axs[0].set_xlabel(r'$Bulk$ ' + r'$Al_2O_3$')
                axs[0].set_ylabel(r'$Melt$ ' + r'$Al_2O_3$')

                axs[1].set_xlim(5, 27)
                axs[1].set_ylim(5, 20)
                axs[1].set_xlabel(r'$Bulk$ ' + r'$Al_2O_3$')
                axs[1].set_ylabel(r'$Melt$ ' + r'$Al_2O_3$')

            elif b == 'MGO(WT%)':
                axs[0].set_xlim(0, 12)
                axs[0].set_ylim(0, 7)
                axs[0].set_xlabel(r'$Bulk$ ' + r'$MgO$')
                axs[0].set_ylabel(r'$Melt$ ' + r'$MgO$')

                axs[1].set_xlim(0, 15)
                axs[1].set_ylim(0, 6)
                axs[1].set_xlabel(r'$Bulk$ ' + r'$MgO$')
                axs[1].set_ylabel(r'$Melt$ ' + r'$MgO$')

            elif b == 'CAO(WT%)':
                axs[0].set_xlim(0, 15)
                axs[0].set_ylim(0, 12.5)
                axs[0].set_xlabel(r'$Bulk$ ' + r'$CaO$')
                axs[0].set_ylabel(r'$Melt$ ' + r'$CaO$')

                axs[1].set_xlim(0, 17.5)
                axs[1].set_ylim(0, 15)
                axs[1].set_xlabel(r'$Bulk$ ' + r'$CaO$')
                axs[1].set_ylabel(r'$Melt$ ' + r'$CaO$')

            elif b == 'K2O(WT%)':
                axs[0].set_xlim(0, 6)
                axs[0].set_ylim(0, 6)
                axs[0].set_xlabel(r'$Bulk$ ' + r'$K_2O$')
                axs[0].set_ylabel(r'$Melt$ ' + r'$K_2O$')

                axs[1].set_xlim(0, 6)
                axs[1].set_ylim(0, 6)
                axs[1].set_xlabel(r'$Bulk$ ' + r'$K_2O$')
                axs[1].set_ylabel(r'$Melt$ ' + r'$K_2O$')
            elif b == 'FEOT(WT%)':
                axs[0].set_xlim(0, 12)
                axs[0].set_ylim(0, 12)
                axs[0].set_xlabel(r'$Bulk$ ' + r'$FeO_t$')
                axs[0].set_ylabel(r'$Melt$ ' + r'$FeO_t$')

                axs[1].set_xlim(0, 17.5)
                axs[1].set_ylim(0, 10)
                axs[1].set_xlabel(r'$Bulk$ ' + r'$FeO_t$')
                axs[1].set_ylabel(r'$Melt$ ' + r'$FeO_t$')
            plt.savefig('./figures/' + m + '_bulk_melt' '.pdf',
                        format="pdf", transparent=True)
            plt.show()
```

```python
fig, axs = plt.subplots()
sns.histplot(vol.mSiO23[vol.mSiO23 <= 80], color='#E60000',
             bins=np.arange(40, 81, 1), stat='probability',)
sns.histplot(plu.mSiO23[plu.mSiO23 <= 80], color='#003153',
             bins=np.arange(40, 81, 1), stat='probability')
axs.set_xlabel(r'$SiO_2$')
axs.set_ylabel(r'$Probability$')
axs.legend(labels=['Volcanic', "Plutonic"])
plt.savefig('./figures/melts_hist.pdf', format="pdf", transparent=True)
plt.show()
```

```python
cal_melt_elm = plu.columns[-17: -3]
for e in cal_melt_elm:
    fig, axs = plt.subplots()
    elm_max = 100
    elm_min = 0
    if e == 'mFeO1':
        elm_max = 11
        axs.set_xlim(0, 11)
        axs.set_xlabel('$FeO_t$')
    if e == 'mMgO':
        elm_max = 8
        axs.set_xlim(0, 8)
        axs.set_xlabel('$MgO$')
    elif e == 'mCaO1':
        elm_max = 11
        axs.set_xlim(0, 11)
        axs.set_xlabel('$CaO$')
    elif e == 'mAl2O3':
        elm_max = 22
        elm_min = 10
        axs.set_xlim(10, 22)
        axs.set_xlabel('$Al_2O_3$')
    elif e == 'mK2O1':
        elm_max = 6
        axs.set_xlim(0, 6)
        axs.set_xlabel('$K_2O$')
    elif e == 'mTiO21':
        elm_max = 1.5
        axs.set_xlim(0, 1.5)
        axs.set_xlabel('$TiO_2$')
    sns.histplot(vol[e][(vol[e] <= elm_max) & (vol[e] >= elm_min)],
                 color='#E60000', bins=40, stat='probability')
    sns.histplot(plu[e][(plu[e] <= elm_max) & (plu[e] >= elm_min)],
                 color='#003153', bins=40, stat='probability')
    axs.set_ylabel(r'$Probability$')
    axs.legend(labels=['Volcanic', "Plutonic"])
    plt.savefig('./figures/melts_hist_' + e + '.pdf',
                format="pdf", transparent=True)
    plt.show()
```
